```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams.update({
    'font.size': 10,
    'font.family': 'serif',
    'axes.labelsize': 10,
    'axes.titlesize': 11,
    'legend.fontsize': 9,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'figure.dpi': 300
})

sns.set_style("whitegrid")
```

```python
import pandas as pd
gym = pd.read_csv('/content/drive/MyDrive/Dataset/gym_membership.csv')
```

```python
plt.figure(figsize=(4,3))

counts = gym['abonoment_type'].value_counts()
counts.plot(kind='bar', edgecolor='black')

plt.xlabel('Membership Type')
plt.ylabel('Number of Members')
plt.title('Distribution of Membership Types')

plt.tight_layout()
plt.savefig('fig1_membership_distribution.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(4,3))

plt.hist(gym['avg_time_in_gym'], bins=12, edgecolor='black')

plt.xlabel('Average Time in Gym (hours)')
plt.ylabel('Frequency')
plt.title('Distribution of Average Gym Time')

plt.tight_layout()
plt.savefig('fig2_avg_time_hist.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(4,3))

plt.scatter(gym['Age'], gym['visit_per_week'], alpha=0.7)

plt.xlabel('Age')
plt.ylabel('Visits per Week')
plt.title('Age vs Weekly Gym Visits')

plt.tight_layout()
plt.savefig('fig3_scatter_age_visits.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(4,3))

sns.boxplot(x='gender', y='Age', data=gym)

plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age Distribution by Gender')

plt.tight_layout()
plt.savefig('fig4_box_gender_age.png', dpi=300)
plt.show()
```

```python
sns.countplot(x='days_per_week', data=gym)
plt.title('Gym Attendance Days per Week')
plt.show()
```

```python
plt.figure(figsize=(4,3))

sns.barplot(x='abonoment_type', y='visit_per_week', data=gym)

plt.xlabel('Membership Type')
plt.ylabel('Average Visits per Week')
plt.title('Visits per Week by Membership Type')

plt.tight_layout()
plt.savefig('fig6_bar_visits_membership.png', dpi=300)
plt.show()
```
