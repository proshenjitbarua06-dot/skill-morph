#H1 Write a function to calculate the sum of numbers from 1 to N.

```python
def calculate_sum(n):
 # Using the arithmetic series formula
    total = n * (n + 1) // 2
    return total
# Example usage:
n_value = 10
result = calculate_sum(n_value)
print(f"The sum of numbers from 1 to {n_value} is: {result}")
```

#H2 Write a function that returns the square of a number.

```python
def square_number(num):
    """Returns the square of a given number."""
    return num ** 2

# Example usage:
value = 5
result = square_number(value)

print(f"The square of {value} is: {result}")
```

#H3 Write a function that checks whether a number is even/odd.

```python
def check_even_odd(number):
    if number % 2 == 0:
        return f"{number} is Even"
    else:
        return f"{number} is Odd"

# Example Usage:
print(check_even_odd(10)) # Output: 10 is Even
print(check_even_odd(7))  # Output: 7 is Odd
```

#H4 Write a function that returns the larger number from two given numbers.

```python
def find_larger(a, b):
    if a > b:
        return a
    else:
        return b

# Example:
print(find_larger(15, 27)) # Output: 27
```

#H6 Write a program to find all unique elements in the list.

```python
def get_unique_elements(data):
    return list(set(data))

# Example Usage:
my_list = [1, 2, 2, 3, 4, 4, 4, 5]
print(get_unique_elements(my_list))
# Output: [1, 2, 3, 4, 5]
```

#H7 Write a Python program to calculate the mean of a list of numbers.

```python
def calculate_mean(numbers):
    if not numbers:
        return 0
    return sum(numbers) / len(numbers)

# Example Usage:
data = [10, 20, 30, 40, 50]
print(f"Mean: {calculate_mean(data)}") # Output: 30.0
```

H8 Classify Numbers into Even and Odd Lists

```python
def classify_numbers(numbers):
    even_numbers = []
    odd_numbers = []

    for num in numbers:
        if num % 2 == 0:
            even_numbers.append(num)
        else:
            odd_numbers.append(num)

    return even_numbers, odd_numbers

# Example Usage:
nums = [12, 7, 21, 18, 5, 2, 44, 13]
evens, odds = classify_numbers(nums)

print(f"Evens: {evens}") # Output: [12, 18, 2, 44]
print(f"Odds:  {odds}")  # Output: [7, 21, 5, 13]
```
